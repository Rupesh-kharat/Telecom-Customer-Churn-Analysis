```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0")
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

#Converted seniorCitizen value to yes pr no instead of 0&1

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python

# Create a normalized crosstab
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Custom colors: blue for No, orange for Yes (adjust order if needed)
colors = ['#1f77b4', '#ff7f0e']  # Adjust depending on Churn label order

# Plot
ax = ct.plot(kind='bar', stacked=True, figsize=(4, 4), color=colors)

plt.title("Churn Percentage by SeniorCitizen", fontsize=10)
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title='Churn', loc='upper right')
plt.xticks(rotation=0)

# Add annotations
for i, row in enumerate(ct.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, cumulative + val / 2, f'{val:.1f}%', ha='center', va='center', fontsize=8, color='white')
        cumulative += val

plt.tight_layout()
plt.show()
```

```python
# Columns to plot
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Color palette matching your screenshot
palette = {'No': '#1f77b4', 'Yes': '#ff7f0e'}

# Set up subplots
fig, axes = plt.subplots(3, 3, figsize=(20, 20))
fig.suptitle('Service Usage by Churn Status', fontsize= 16)

axes = axes.flatten()

# Plot with hue and color palette
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue='Churn', palette=palette, ax=axes[i])
    axes[i].set_title(f'Count of {col}', fontsize= 16)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].legend(title='Churn')

# Remove extra axes if needed
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```

#The visualizations highlight churn trends across telecom service features. Customers who **lack services** such as OnlineSecurity, TechSupport, or DeviceProtection exhibit **higher churn rates**. In contrast, those with active subscriptions to these features tend to stay longer. Notably, users with **no internet-related services** or limited engagement (e.g., no streaming) are also more prone to churn, indicating service usage plays a key role in customer retention.
